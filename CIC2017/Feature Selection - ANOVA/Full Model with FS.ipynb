```python
import numpy as np
import pandas as pd

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

import math
from numpy import inf
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("CIC2017.csv")
```

```python
df
```

```python
df.describe()
```

```python
df = df[~df.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
X = df.drop(['_label'], axis=1)
y = df['_label']

x_columns = X.columns
```

```python
sc = StandardScaler()

X = np.array(X, dtype=float)
X[X == inf] = 9999
X = sc.fit_transform(X)
print(np.isnan(X).any())
X=np.nan_to_num(X)

X = pd.DataFrame(X)
X.columns = x_columns
```

```python
train_X = pd.DataFrame(X)
train_y = pd.DataFrame(y)
output = pd.concat([train_X, train_y], axis=1, sort=False)

output = output.dropna()
```

```python
x_columns = pd.Series(x_columns)
trial = x_columns.reset_index()
trial = trial[0]
trial = trial.values.tolist()
trial.append('_label')
```

```python
output.columns = trial
output = output.reset_index()

del output['index']
```

```python
# Select all attacks from the dataset
attacks = output.loc[(output['_label'] != 'BENIGN')]
attacks = attacks.sample(frac=0.01)
attack_dim = len(attacks.index)

benign = output.loc[(output['_label'] == 'BENIGN')]
benign = benign.sample(frac=0.15)
normal_dim = len(benign.index)

file = pd.concat([attacks, benign])

print(attacks.shape)
file.sort_index(inplace=True)
file
```

```python
X = file.drop(['_label'], axis=1)
y = file['_label']
```

```python
y = y.replace(to_replace = 'BENIGN', value = 0)
y = y.replace(to_replace = 'DDoS', value = 1)
```

```python
y.unique()
```

```python
# define feature selection
fs = SelectKBest(score_func=f_classif, k=10)
```

```python
# apply feature selection
X_selected = fs.fit_transform(X, y)
print(X_selected.shape)
X_selected
```

```python
mask = fs.get_support() # List of booleans
new_features = [] # The list of your K best features

for bool, feature in zip(mask, x_columns):
    if bool:
        new_features.append(feature)
```

```python
new_features
```

```python
''' Best 10 Features:
['_Destination_Port',
 'Bwd_Packet_Length_Max',
 '_Bwd_Packet_Length_Mean',
 '_Bwd_Packet_Length_Std',
 '_Min_Packet_Length',
 '_Packet_Length_Mean',
 '_Packet_Length_Std',
 '_URG_Flag_Count',
 '_Average_Packet_Size',
 '_Avg_Bwd_Segment_Size']
 '''
```

```python
X = X[new_features]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import fbeta_score, make_scorer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```

```python
# Create Learners per layer
layer_one_estimators = [
                        ('knn_1', KNeighborsClassifier(n_neighbors=5, n_jobs=-1)),
                        ('mlp_1', MLPClassifier(solver='adam', alpha=0.001, hidden_layer_sizes=20))
                        ]

layer_two_estimators = [
                        ('svm_2', SVC(decision_function_shape='ovo')),
                        ('lr_2', LogisticRegression())
                       ]
```

```python
layer_two = StackingClassifier(estimators=layer_two_estimators, final_estimator=DecisionTreeClassifier())
```

```python
import time

# Create Final model by 
clf = StackingClassifier(estimators=layer_one_estimators, final_estimator=layer_two)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

start_time = time.time()
clf.fit(X_train, y_train)
train_time = time.time() - start_time
print("--- %s seconds ---" % train_time)

y_pred = clf.predict(X_test)

acurcy = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
recal = recall_score(y_test, y_pred)
```

```python
clf_Score = pd.DataFrame(columns=['Accuracy', 'Precision', 'Recall', '[No. Features]', '[Time Taken for Training]',
                                 '[Class Balance - Attack/Normal]'])
clf_Score.at[0, 'Accuracy'] = acurcy
clf_Score.at[0, 'Precision'] = prec
clf_Score.at[0, 'Recall'] = recal
clf_Score.at[0, '[No. Features]'] = len(new_features)
clf_Score.at[0, '[Time Taken for Training]'] = train_time
clf_Score.at[0, '[Class Balance - Attack/Normal]'] = attack_dim/normal_dim
print(clf_Score)
```

```python
clf_Score.to_csv(str(len(new_features)) + '-CIC_DOS_Stacking_level2_(KNN-MLP)_(SVM-LR)_DT.csv')
```
